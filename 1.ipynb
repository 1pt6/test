```python
import pandas as pd  

# 读取CSV文件，假设文件名为data.csv，并且需要处理的sheet名为sheet2  
file_path = 'data.xlsx'  
sheet_name = '优秀学生奖学金'  

# 读取Excel文件中的特定sheet到DataFrame  
# 注意：如果你的文件是CSV格式，通常不会有多个sheet，但这里我们假设你的文件是以某种方式包含多个sheet（例如实际是Excel文件）  
# 如果是CSV文件，直接读取整个文件即可，不需要指定sheet_name  
# df = pd.read_csv(file_path)  # 直接读取CSV文件  

# 由于CSV通常不包含多个sheet，这里假设是Excel文件，如果是Excel文件，请使用以下代码读取特定sheet  
# 需要安装 openpyxl 或 xlrd 库来读取Excel文件  
# pip install openpyxl xlrd  
if file_path.endswith('.csv'):  
    df = pd.read_excel(file_path)  # 直接读取CSV  
elif file_path.endswith(('.xls', '.xlsx')):  
    xl = pd.ExcelFile(file_path)  
    df = xl.parse(sheet_name)  # 读取指定sheet  
else:  
    raise ValueError("Unsupported file format")  

# 检查数据  
print(df.head())  

# 按“获奖名称”列进行分类并统计总数  
grouped = df.groupby('获奖名称').size().reset_index(name='总数')  

# 打印结果  
print(grouped)  

# 如果需要将结果保存到新的CSV文件  
# grouped.to_csv('获奖名称分类统计.csv', index=False)
```
